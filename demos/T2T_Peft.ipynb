```python
import text2text as t2t
from huggingface_hub import notebook_login
notebook_login()
```

```python
peft_trainer = t2t.SFTuner()

model_name = "microsoft/phi-1_5"
peft_trainer.prepare_model(model_name) # Prepares model for PEFT, and tokenizer
```

```python
# Prepare dataset

peft_trainer.prepare_dataset() # Default finetuning on smangrul/ultrachat-10k-chatml
```

```python
## Train

peft_trainer.train(
    max_steps = 100,
    learning_rate = 2e-4,
    lr_scheduler_type = "linear",
    per_device_train_batch_size = 2,
    per_device_eval_batch_size = 2,
    gradient_accumulation_steps = 4,
    warmup_steps = 5,
    logging_steps = 20,
    eval_strategy = "steps",
    fp16 = True,
    bf16 = False,
    )
```
